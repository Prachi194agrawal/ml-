## seaborn distribution plot

## distplot= histogram +kdeplot+rugplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('white')
```

## Intro Visuals

## Basics

```python
sns.distplot(cars.weight);
```

```python
sns.distplot(cars.weight,kde=False);
```

```python
sns.distplot(cars.weight,hist=False,rug=True);
```

## options

## bins

```python
sns.distplot(cars.weight,bins=20);
```

```python
sns.distplot(cars.weight,bins=[1500,2000,2500,3500,5200]);
```

## fit

```python
from scipy.stats import norm

sns.distplot(cars.weight,fit=norm);
```

## Kwargs

# hist_kws

```python
sns.distplot(cars.weight,kde=False);
```

```python
sns.distplot(cars.weight,kde=False,hist_kws={'alpha':1});
```

```python
sns.distplot(cars.weight,kde=False,
            hist_kws={'alpha':1,'histtype':'step'});
```

```python
sns.distplot(cars.weight,kde=False,
            hist_kws={'alpha':1,'histtype':'stepfilled'});
```

## kde_kws

```python
sns.distplot(cars.weight);
```

```python
sns.distplot(cars.weight,kde_kws={'lw':5});
```

```python
sns.kdeplot(cars.weight,lw=5);
```

```python
sns.distplot(cars.weight,kde_kws={'bw':1});
```

## rug_kws

```python
sns.distplot(cars.weight,rug=True,rug_kws={'height':0.15});
```

```python
sns.distplot(cars.weight,rug=True,color='r',
            rug_kws={'height':0.15});
```

```python
sns.distplot(cars.weight,rug=True,
            rug_kws={'height':0.15,'color':'xkcd:navy'});
```
