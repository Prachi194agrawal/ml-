## seaborn heatmap

```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
cars=sns.load_dataset('mpg')
```

```python
cars.head()
```

## basics

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
(cars
.groupby('origin')
.cylinders
.value_counts()
.unstack())
```

```python
origin_cyl=(
cars
.groupby('origin')
.cylinders
.value_counts()
.unstack()
.fillna(0))
```

```python
sns.heatmap(origin_cyl);
```

## transpose

```python
# Assuming 'cars' is your DataFrame
encoded_cars = pd.get_dummies(cars, columns=['origin'])
correlation_matrix = encoded_cars.corr()
sns.heatmap(correlation_matrix)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'cars' is your DataFrame
numeric_cars = cars.select_dtypes(include=['number'])
correlation_matrix = numeric_cars.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

## color palette

seaborn offers 170 different color palette options ,check out this great post to view them all.
We will start with a SEQUENTIAL palette.

```python
sns.heatmap(origin_cyl,cmap='Blues');
```

color diverging

```python
sns.heatmap(numeric_cars.corr(),cmap='RdBu',center=0);
```

## annoations

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True);
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,fmt=" .1f");
```

```python
sns.heatmap(origin_cyl,
           cmap='Blues',
           annot=True,
           fmt=" .0f",
           annot_kws={
               'fontsize':16,
               'fontweight':'bold',
               'fontfamily':'serif',
               'color':'black',
           });
```

```python
## advanced styling
sns.heatmap(numeric_cars.corr(),cmap='RdBu',annot=True,linewidth=1);
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,linewidth=1);
```

```python
sns.heatmap(origin_cyl,cmap='Blues',annot=True,linewidth=1,linecolor='black');
```

```python
labels=['prs','ytr','yui']
```

```python
sns.heatmap(numeric_cars.corr(),cmap='RdBu',
            xticklabels= False,
           yticklables=labels);
```

```python
sns.heatmap(origin_cyl,cmap='Blues',
            xticklabels= labels,
           yticklables=labels);
```
