## box plot components

## seaborn :bornplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

## intro visuals

## basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

filter down to cars with even number of cylinders

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.boxplot(cars.mpg);
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin,y=cars.mpg);
```

```python
sns.boxplot(x='origin',y='mpg',data=cars);
```

```python
sns.boxplot(x='origin',y='mpg',hue='cylinders',data=cars);
```

## by model year

```python
cars.model_year.describe()
```

```python
cars['newer_model']=cars.model_year>76
```

```python
sns.boxplot(x='origin',y='mpg',hue='newer_model',data=cars);
```

## styling options

## order

```python
sns.boxplot(x='origin',y='mpg',
           data=cars
           );
```

```python
sns.boxplot(x='mpg',y='origin',
           data=cars
           );
```

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           order=['japan','europe','usa']
           );
```

```python
sns.boxplot(x='mpg',y='origin',hue='newer_model',
           data=cars,
           order=['japan','europe','usa'],
            hue_order=[True,False]
           );
```

## color

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           color='r'
           );
```

```python
sns.boxplot(x='mpg',y='origin',hue='newer_model',
           data=cars,
           color='r'
           );
```

## width

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           width=0.5
           );
```

## linewidth

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           linewidth=2.5
           );
```

## whis

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           whis=1
           );
```

## fliersize

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           fliersize=20
           );
```

## showcaps

```python
sns.boxplot(x='mpg',y='origin',
           data=cars,
           showcaps=False
           );
```
