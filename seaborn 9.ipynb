## scatterplot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds=sns.load_dataset('diamonds')
diamonds.shape
```

```python
diamonds.head()
```

```python
diamonds=diamonds[
    diamonds.cut.isin(['Premium','Good'])&
    diamonds.color.isin(['D','F','J'])
].sample(n=100,random_state=22)
diamonds.shape
```

## intro visiuals

## basics

```python
sns.set_style('dark')
```

```python
sns.scatterplot(x=diamonds.carat,y=diamonds.price);
```

```python
sns.scatterplot(x='carat',y='price',data=diamonds)
```

```python
sns.scatterplot(x='x',y='y',data=diamonds);
```

## sementic variables

```python
sns.set_style('darkgrid')
```

## hue

```python
sns.scatterplot(x='carat',y='price',hue='cut',
               data=diamonds
               );
```

```python
sns.scatterplot(x='carat',y='price',hue='cut',
               data=diamonds,
               #palette=['red','#55CCCC']
               palette='summer');
```

## continuous variable

```python
sns.scatterplot(x='carat',y='price',hue='depth',
               data=diamonds
               );
```

## style

```python
sns.scatterplot(x='carat',y='price',style='cut',
               data=diamonds,
               s=100);
```

## size

```python
sns.scatterplot(x='carat',y='price',size='cut',
               data=diamonds);
```

```python
sns.scatterplot(x='carat',y='price',size='cut',
               data=diamonds,
               sizes=[150,50]);
```

```python
sns.scatterplot(x='carat',y='price',size='depth',
               data=diamonds);
```

## hue+style+size

```python
sns.scatterplot(x='carat',y='price',hue='cut',style='color'
               ,data=diamonds,s=100);
```

## alpha

```python
sns.scatterplot(x='carat',y='price',
               data=diamonds,s=100,alpha=0.6);
```

## marking style

```python
sns.scatterplot(x='carat',y='price',
               data=diamonds,s=100,
               marker='*',edgecolor='black');
```
