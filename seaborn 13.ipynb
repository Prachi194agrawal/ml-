## pairplot

```python
import seaborn as sns
from matplotlib import pyplot as plt 
import numpy as np
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

## intro visuals

##### basics

```python
sns.set_style('darkgrid')
```

```python
sns.pairplot(tips);
```

```python
tips.dtypes
```

note that the axes tick labels correspond to the bivariate ,relational plot and Not the distribution plots

```python
sns.distplot(tips['size'],
            kde=False,
            hist_kws={'alpha':1});
```

# plot kinds

## diag-kind

```python
sns.pairplot(tips,diag_kind='kde');
```

## kind

```python
sns.pairplot(tips,kind='reg');
```

## returns pairplot

```python
g=sns.pairplot(tips)
```

```python
type(g)
```

```python
g=sns.pairplot(tips)
g.map_upper(sns.kdeplot,n_levels=6,color='xkcd:wine red');
```

## categorical and specific variables

```python
tips['weekend']=tips.day.isin(['Sat','Sun'])
```

```python
tips.head()
```

##hue

```python
sns.pairplot(tips,hue='weekend');
```

```python
tips.dtypes
```

# vars,x_vars,y_vars

```python
sns.pairplot(tips,hue='weekend',vars=['total_bill','tip']);
```

```python
sns.pairplot(tips,hue='weekend',y_vars=['tip'],x_vars=['total_bill','size'],
            kind='reg',height=4);
```

## styling

```python
del tips['weekend']
```

## height ,aspect

```python
from ipywidgets import interactive
def size_widget(height=2.5,aspect=1):
    sns.pairplot(tips,height=height,aspect=aspect)
```

```python
interactive(size_widget,height=(1,3.5,0.5),aspect=(0.5,2,0.25))
```

## palette

```python
sns.pairplot(tips,hue='day',palette='plasma');
```

## diag_kws,plot_kws

```python
sns.pairplot(tips,diag_kws={'color':'gray','alpha':0.6});
```

```python
sns.pairplot(tips,
            kind='reg',
            plot_kws={'ci':None,'color':'xkcd:salmon',
                     'scatter_kws':{'color':'gray'}});
```
