# seaborn barplot

```python
import seaborn as sns
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
penguins=sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.shape
```

## intro visiuals

## basics

```python
sns.set_style('darkgrid')
```

```python
sns.barplot(x='species',y='body_mass_g',data=penguins);
```

```python
penguins.groupby('species').body_mass_g.mean()
```

## horizontal barplot

```python
sns.barplot(x='body_mass_g',y='species',data=penguins);
```

## interval estimates

## confidence level(ci)

```python
sns.barplot(x='species',y='body_mass_g',
           data=penguins,ci=None);
```

## estimator

```python
sns.barplot(x='species',y='body_mass_g',
           data=penguins,
           estimator=np.std);
```

```python
penguins.groupby('species').body_mass_g.describe()
```

```python
sns.barplot(x='species',y='body_mass_g',
           data=penguins,estimator=np.max);
```

## catorical variable

```python
sns.barplot(x='species',y='body_mass_g',
           data=penguins,
           hue='sex');
```

## order ,hue_order

sns.barplot(x='species',y='body_mass_g',data=penguins);

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,hue='sex',hue_order=['Female','Male'],
           order=['Gentoo','Chinstrap','Adelie']);

```

##styling

## color

```python
sns.barplot(x='species',y='body_mass_g',
           data=penguins,color='c',errcolor='black');
```

## [palette]

```python
sns.barplot(x='species',y='body_mass_g',
           data=penguins,
           palette='summer');
```

## edgecolor and linewidth

the seaborn barplot inherits from matplotlib's bar .Check the docs for more options

```python
sns.barplot(x='species',y='body_mass_g',data=penguins,ci=None,
           color='aliceblue',edgecolor='mediumblue',
           lw=3);
```
