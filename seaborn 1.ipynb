```python
from matplotlib import pyplot as plt
import seaborn  as sns
```

## univariate(1D)

```python
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower,shade=True)
```

```python
sns.kdeplot(cars.horsepower,bw=0.05);
```

```python
sns.kdeplot(cars.horsepower,bw=5);
```

```python
sns.kdeplot(cars.horsepower,cumulative=True);
```

## bivariate intro

## bivariate Demo

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg);
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,n_levels=20);
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,shade=True);
```

```python
sns.set_style('whitegrid')
sns.kdeplot(x=cars.horsepower,y=cars.mpg,shade=True,
           shade_lowest=False);
```

```python
sns.kdeplot(x=cars.horsepower,y=cars.mpg,shade=True,
           shade_lowest=False,cbar=True);
```
