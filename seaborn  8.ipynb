## seaborn stirplot

```python
import seaborn as sns
```

```python
cars=sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

## intro visiuals

## basics

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

filter down to only cars with even number of cylinders

```python
cars=cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.stripplot(cars.weight);
```

```python
sns.stripplot(x=cars.weight,y=cars.origin);
```

```python
sns.stripplot(x='weight',y='origin',hue='cylinders',
             data=cars,palette='summer'
             );
```

```python
sns.stripplot(x='weight',y='origin',hue='cylinders',
             data=cars,
             dodge=True
             );
```

## orientation

```python
sns.stripplot(x='weight',y='origin',data=cars);
```

```python
sns.stripplot(y='weight',x='origin',data=cars);
```

## jitter

```python
sns.stripplot(x=cars.weight,y=cars.origin,jitter=False)
```

```python
sns.stripplot(x=cars.weight,y=cars.origin,jitter=0.05);
```

```python
from ipywidgets import interactive 

def jitter_widget(jitter=0.1):
    sns.stripplot(x=cars.weight,y=cars.origin,jitter=jitter)
```

```python
interactive(jitter_widget,jitter=(0,0.65,0.05))
```

## alpha

```python
sns.set_style('whitegrid');
```

```python
sns.stripplot(x=cars.weight,y=cars.origin,alpha=0.35);
```

```python
sns.stripplot(x=cars.weight,y=cars.origin,
             alpha=0.3,
             size=8
             );
```

## linewidth,edgecolor

```python
sns.stripplot(x=cars.weight,y=cars.origin,linewidth=1);
```

```python
sns.stripplot(x=cars.weight,y=cars.origin,linewidth=1,edgecolor='black');
```

check the matplotlib scatter and the marker style docs for more styling of the markers themselves

```python
sns.stripplot(x=cars.weight,y=cars.origin,
             alpha=0.7,size=10,jitter=0.25,marker='*');
```
